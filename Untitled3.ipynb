Colot threshold: Blue screen

Imports

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
#### Read Image
```

```python
image = cv2.imread('images/pizza_bluescreen.jpg')

image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

print('Image dimensions:', image_rgb.shape)

plt.imshow(image_rgb)
```

```python
space = cv2.imread('images/space_background.jpg')

space_rgb = cv2.cvtColor(space, cv2.COLOR_BGR2RGB)

print('Image dimensions:', space_rgb.shape)

plt.imshow(space_rgb)
```

```python
low_blue = np.array([0, 0, 100])
upper_blue = np.array([80, 80, 255])
```

```python
mask = cv2.inRange(image_rgb, low_blue, upper_blue)

plt.imshow(mask, cmap="gray")
```

```python
masked_image = np.copy(image_rgb)
masked_image[mask!=0] = [0,0,0]

plt.imshow(masked_image)
```

```python
sub_image = space_rgb[0:514, 0:816]
plt.imshow(sub_image, cmap='gray')
```

```python
masked_space = np.copy(sub_image)
masked_space[mask==0] = [0,0,0]

final_image = masked_image + masked_space

plt.imshow(final_image)
```
